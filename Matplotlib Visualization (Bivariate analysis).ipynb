# 1. line plot

      - default  plot( ) present in pyplot

```python
import matplotlib .pyplot as plt
import numpy as np
import pandas as pd
```

```python
l1=[2,3,4]
l2= [4,5,6]
plt.plot(l1,l2)  # or we can use semicolon alse for removing in place of plt.show()
plt.show() 
```

```python
# adding labels
l1=[2,3,4]
l2= [4,5,6]
plt.plot(l1,l2)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot");  
```

```python
# adding the markers
l1=[2,3,4]
l2= [4,5,6]
plt.plot(l1,l2,marker="^",markersize=20)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot");
```

```python
# adding the colors
l1=[2,3,4]
l2= [4,5,6]
plt.plot(l1,l2,color="green",marker="^",linestyle="--",markersize=20)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot");
```

```python
# making a line plot using array
x= np.arange(1,11)
y= np.arange(12,22)
plt.plot(x,y)
plt.title("line graph")
plt.ylabel("y axis")
plt.xlabel("x axis")
```

```python
x= np.arange(1,11)
y= np.arange(12,22)
plt.plot(x,y,color="g",marker="^",markersize=20)
plt.title("line graph")
plt.ylabel("y axis")
plt.xlabel("x axis")
```

```python
x= np.arange(1,11)
y= x**3
plt.plot(x,y,color="g",marker="^",markersize=20)
plt.title("line graph");
plt.ylabel("y axis")
plt.xlabel("x axis")
```

```python
# comparsion : multiple lines in one figure
x1=np.linspace(1,10,5)
x2=np.linspace(10,20,5)
plt.figure(figsize=(5,3))   # 5 is lenghth nd 3 is breadth
x3=x2*2
x4=x2*3
plt.plot(x1,x2,label="2000")
plt.plot(x1,x3,label="2005")
plt.plot(x1,x4,label="2010")
plt.legend()
plt.show()
```

```python
l=[1,2,3,4,5,6,7] ## product id
price=[400,5000,32002,60005,70089,90000,70064] ##  total price
plt.plot(l,price,color="r",marker="*",markersize=10)
plt.xlabel("Product id")
plt.ylabel("product-price")
plt.title("product price comparisions")
plt.show()       ##it is used to remove the information shown by amtplotlib
```

```python
l=[1,2,3,4,5,6,7] ## product id
plt.figure(figsize=(5,3),dpi=200)
online_price=[400,500,322,65,789,90,764]  ##  online price
offline_price=[450,556,400,100,770,110,800] ### offline price
plt.plot(l,online_price,color="r",marker="*",label="onlineprice")
plt.plot(l,offline_price,color="g",marker="^",label="offlineprice")
plt.xlabel("Product id")
plt.ylabel("product-price")
plt.title("product price comparisions")
plt.legend()
```

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.plot(year,online_sale,color="r",marker="*",label="onlineprice")
plt.xticks(rotation=45)
plt.xlabel("sale_year")
plt.ylabel("online_sale")
plt.title("trend of the sale")
plt.legend()

```

# 3. BAR PLOT

# Bar plot
      - Bivariate Analysis
      - when one variable is categorical and other is numeric
      - it gives maximum values

## difference between bar plot and seaborn 
### seaborn gives average value by default but bar plot gives maximum value


```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']   ## categorical
students = [23,30,35,40,12]   ## numeric
## to know in which course we have maximum no. of studentsp
plt.bar(langs,students)
```

```python
fig = plt.figure()
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.bar(langs,students,color ='green',width = 0.6)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()


```

```python
# compare year wise sale
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.bar(year,online_sale,color="k")
plt.xticks
plt.show()
```

```python
# eg.2to know which player got maximum score card
from matplotlib import pyplot as plt  
players = np.array(['Virat','Rohit','Shikhar','Hardik']) ## player names
runs = np.array([51,87,45,97])   ## runs
plt.bar(players,runs,color = 'yellow',width=0.49)  
plt.title('Score Card')  
plt.xlabel('Players')  
plt.ylabel('Runs')  
plt.show()  

```
